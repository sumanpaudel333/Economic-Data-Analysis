The task for this assignment primarily focuses on analyzing the economic data of different countries, comparing economic indicators (GDP, unemployment, and growth) with inflation using various models. The dataset is taken from the online source and may contain redundant data as well. So, we first go through the different steps from data cleaning to processing using different models and choose the best one that fits the available data.
Now, the requirement of this task is outlined below:
1. Whether the economic indicators most strongly influence inflation or not.
2. Can inflation be really predicted using the indicators like GDP, Unemployment and growth?
3. Which regression model provides the most accurate predictions for this dataset?
Let's move through the data workflow and analyse the available data.

Firstly, let's import all the required Python libraries which we are going to use later in different steps of data processing, visualization, modeling, and evaluation.

```python
# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing and model selection
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import classification_report, accuracy_score

# Regression models
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans

# Evaluation metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

Now, the first part comes to the data loading and preprocessing of the dataset. Here we load the dataset named `Economic Indicators And Inflation.csv`, clean column names, select relevant features, and scale the data for modeling.

```python
# Load the dataset
df = pd.read_csv("Economic Indicators And Inflation.csv")

df.columns = df.columns.str.strip()

# Rename columns for easier access (if not already renamed)
df.rename(columns={
    'GDP (in billion USD)': 'GDP',
    'Inflation Rate (%)': 'Inflation',
    'Unemployment Rate (%)': 'Unemployment',
    'Economic Growth (%)': 'Growth'
}, inplace=True)

# Filter the dataset to recent years
df_filtered = df[df['Year'] >= 2022].copy()

# Keep only relevant columns
df = df[['GDP', 'Inflation', 'Unemployment', 'Growth']].dropna()

# Define features and target
X = df[['GDP', 'Unemployment', 'Growth']]
y = df['Inflation']

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize Linear Regression model
model = LinearRegression()

# Perform RFE
rfe = RFE(estimator=model, n_features_to_select=2)
rfe.fit(X_train, y_train)

# Get selected features
selected_features = X.columns[rfe.support_]
ranking = rfe.ranking_

# Print results
print("Selected Features:", list(selected_features))
print("Feature Ranking:", dict(zip(X.columns, ranking)))
df.head()
```

Alright, we have done the data preprocessing and loaded the data, along with performing Recursive Feature Elimination(RFE) to select the most relevant features for predicting inflation.
Now, in this step, we move on to check for the missing values and clean up the data accordingly.

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Drop rows with missing values (you could also impute if necessary)
df.dropna(inplace=True)

# Check for duplicates and drop them
df.drop_duplicates(inplace=True)

# Reset index after dropping rows
df.reset_index(drop=True, inplace=True)
```

Fortunately, there were not much of the missing values in the dataset. Now, we define X and Y from the filtered dataset and standarize the feature through standard scalar.

```python
# Define X and y from filtered dataset
X = df_filtered[['GDP', 'Unemployment', 'Growth']]
y = df_filtered['Inflation']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

In the next step, we split the data into a train and test dataset and train from Linear Regression model as a baseline and evaluate it through the performance metrics, whether the model fits well in the given data or not. We use the Mean Squared Error (MSE), Mean Absolute Error(MAE), and R-squared (R²) metrics to evaluate the result.

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train baseline Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict on test data
y_pred_lr = lr_model.predict(X_test)

# Evaluate the model
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

# Print performance metrics
print(f"Linear Regression Results:")
print(f"MAE: {mae_lr:.3f}")
print(f"MSE: {mse_lr:.3f}")
print(f"R²:  {r2_lr:.3f}")
```

Whooo! The result is not quite as expected. 
It suggests that the linear model explains only about 19.7% of the variance in inflation based on the selected features(GDP, Unemployment, and Growth)
and the low R² value indicates that a simple linear relationship might not be sufficient to capture the underlying patterns in the data.

```python
# Use PolynomialFeatures to transform input data
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_scaled)

# Train-test split
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# Train model
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train_poly)

# Predict and evaluate
y_pred_poly = poly_model.predict(X_test_poly)

mae_poly = mean_absolute_error(y_test_poly, y_pred_poly)
mse_poly = mean_squared_error(y_test_poly, y_pred_poly)
r2_poly = r2_score(y_test_poly, y_pred_poly)

print("Polynomial Regression (Degree 2) Results:")
print(f"MAE: {mae_poly:.3f}")
print(f"MSE: {mse_poly:.3f}")
print(f"R²:  {r2_poly:.3f}")
```

After applying Polynomial Regression with degree 2, we observed a significant improvement in model performance:
This means the model now explains approximately 70.2% of the variance in inflation, indicating a much better fit compared to the linear model. The decrease in error values (MAE and MSE) confirms that the polynomial model captures the non-linear relationships among economic indicators more effectively.
Based on these results, we proceed with Polynomial Regression as our selected model and now move on to visualize the predicted vs actual inflation values to better understand the model's performance and how well it fits the data.

```python
# Plot: Actual vs Predicted Inflation (Polynomial Regression)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test_poly, y=y_pred_poly)
plt.plot([y_test_poly.min(), y_test_poly.max()],
         [y_test_poly.min(), y_test_poly.max()],
         color='red', lw=2, linestyle='--', label='Perfect Fit')

plt.title('Actual vs Predicted Inflation (Polynomial Regression)')
plt.xlabel('Actual Inflation')
plt.ylabel('Predicted Inflation')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

sns.pairplot(df)
plt.suptitle("Pairwise Plots of Features", y=1.02)
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(df['Inflation'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Inflation')
plt.xlabel('Inflation Rate (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=numeric_df)
plt.title("Boxplot for Outlier Detection")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Different results can be drawn from the generated graphical representations above. Some of the analyses done through the visuals can be listed as:
1. The scatter plot compares the actual inflation values(x-axis) to the predicted values(y-axis). Most of the data points are clustered near the perfect fit line(red line), which shows the perfect fit of the model. However, there are some outliers as well, which can be seen clearly in the graph.
2. The heatmap shows different correlation values which can be analysed as the range from 1(strong negative correlation) to 1(strong positive correlation). Inflation has a weak positive correlation with Unemployment and growth, and GDP shows a negative correlation with other values. Most of the correlations are weak, which suggests that no single feature is dominant and the combination of multiple features should be considered for prediction.
3. The pair-wise plot shows different correlations of variables with each other. Some of the noted results are:
       a. GDP and unemployment have a notable negative correlation, i.e Higher the GDP, the  Lower the unemployment.
       b. Growth and unemployment also appear inversely related.
       c. Inflation shows a weaker or no strong correlation with other variables.
       d. Data include some outliers particularly in GDP and inflation.
4. The inflation data is highly concentrated in the low range (0–10%), with a few extreme outliers causing a long tail to the right.

```python
inertia = []
K_range = range(1, 10)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
# Apply KMeans with optimal k
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add the cluster labels
df_filtered['Cluster'] = clusters

country_cluster_unique = df_filtered.groupby('Country')['Cluster'] \
                           .agg(lambda x: x.value_counts().index[0]) \
                           .reset_index()

# Rename for clarity
country_cluster_unique.columns = ['Country', 'Most Frequent Cluster']

# Display the result
print(country_cluster_unique.to_string(index=False))
# Visualize clusters
sns.pairplot(df_filtered, hue='Cluster', vars=['GDP', 'Unemployment', 'Growth', 'Inflation'], palette='Set1')
plt.suptitle("KMeans Clusters of Countries", y=1.02)
plt.show()
```

A different conclusion can be drawn after the representation of clusters in the above pairplot. We took k=4 for the clustering and the visualization is separated as well. Some of the results drawn from the figure can be listed as:
1. Cluster-0 (Red): Low GDP, moderate unemployment, variable growth,, and moderate inflation.
2. Cluster-1 (Blue): Moderate GDP, high unemployment, low growth, and moderate inflation — potentially stagnant economies.
3. Cluster-2 (Green): High GDP, low unemployment, high growth, and moderate inflation — likely the most economically stable cluster.
4. Cluster-3 (Purple): Low GDP, moderate unemployment, moderate to high growth, and variable inflation — potentially volatile or emerging economies.
